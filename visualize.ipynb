```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
eltp_train_features = "extracted_features/aldp_features/aldp_train.csv"
df = pd.read_csv(eltp_train_features)
df
```

```python
# count bonafide
bonafide_count = sum(df['KEY']=="bonafide")
bonafide_count
```

```python
plt.figure(figsize=(20,12))
sns.lineplot(df[df.columns[6:]])
ax = plt.gca()
x = np.linspace(0,bonafide_count,bonafide_count+1)
y1 = [0] * len(x)
y2 = [0.0065] * len(x)
ax.fill_between(x, y1, y2, alpha=.5, )
ax.text(0.6, 0.006, "Bonafide")
plt.savefig('plots/feature_line_plot_aldp.png')
```

```python
for i in range(1,21):
    feature_name = f"FEATURE_{i}" 
    plt.figure(figsize=(10, 6))
    sns.boxplot(df[df.columns[5:]], y =feature_name , x="KEY")
    plt.title(feature_name + " Box Plot for Bonafide and Spoof")
    plt.savefig(f'plots/box_plot/aldp_features/feature_{i}_box_plot_aldp.png')
```

```python
for i in range(1,21):
    feature_name = f"FEATURE_{i}" 
    plt.figure(figsize=(10, 6))
    sns.histplot(df[df.columns[5:]], x =feature_name, hue="KEY", kde=True, element='step')
    plt.title(feature_name + " Box Plot for Bonafide and Spoof")
    plt.savefig(f'plots/histogram_plot/aldp_features/feature_{i}_histogram_plot_aldp.png')
```
